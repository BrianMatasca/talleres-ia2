# Agente de recomendación de posgrados Icesi

El Recomendador de Posgrados Icesi es un agente conversacional diseñado para orientar a estudiantes y profesionales en la elección de un posgrado en la Universidad Icesi. Utiliza como base de conocimiento un archivo Excel con todos los programas disponibles (maestrías, especializaciones y doctorados), y a partir de los intereses del usuario (área de estudio, modalidad, duración, presupuesto, competencias de salida) entrega recomendaciones personalizadas sin necesidad de buscar manualmente en catálogos extensos.

### Pasos:

En la primera parte creé mi agente llamado “Agent Recomendación Posgrados” y le escribí una descripción clara para explicar que sirve como recomendador de programas de maestría, especialización y doctorado en la Universidad Icesi, detallando que su función es dar recomendaciones personalizadas según intereses, modalidad, créditos, duración y competencias.

image.png

Después subí el archivo programas_posgrados.xlsx, que se convirtió en la base de conocimiento principal del agente, y confirmé que quedara agregado para que toda la información de los programas se tomara de allí.

image.png

Luego configuré las instrucciones del agente, donde indiqué que debía llamarse Recomendador de Posgrados Icesi y que usara únicamente los datos del Excel cargado, teniendo en cuenta columnas como programa, duración, créditos, modalidad, valor, descripción y competencias de salida.

image.png

Más adelante diseñé el flujo del tema “Recomendar posgrado”, que comienza con un mensaje de bienvenida al usuario y sigue con una pregunta para identificar el área de interés, como Medicina, Ingeniería o Negocios.

image.png

Finalmente, probé el agente escribiendo “Recomiéndame un posgrado”, y este respondió correctamente saludando y pidiendo más detalles sobre los intereses del usuario, comprobando que el flujo funcionaba como lo había diseñado.
